3. Selecting Data

```python
# Access a specific column
# This retrieves the data from the specified column in the DataFrame.
# Replace 'ColumnName' with the actual name of the column you want to access

column_data = df['Gender'] # Accesses the column named 'ColumnName'
column_data
# Access rows by index position
# This retrieves the first row of the DataFrame using its indexposition (0-based).
first_row = df.iloc[0] # Accesses the first row (index 0)
# Access rows by index label
# This retrieves the row using its index label.
# Ensure that the label exists in the DataFrame's index.
first_row_label = df.loc[0] # Accesses the row with index label 0
print(df)
```

4. Filtering Data

```python
filtered_data = df[df['Attendance_Rate'] > 10]
filtered_data
```

5. Adding and Removing Columns

```python
# Add a new column
# This creates a new column 'NewColumn' in the DataFrame by adding the values from 'Column1' and 'Column2'.
# Replace 'Column1' and 'Column2' with the actual column names you want to use for the addition.
df['NewColumn'] = df['Past_Exam_Scores'] + df['Final_Exam_Score'] # Creates a new column based on the sum of two existing columns
# Remove a column
# This removes the specified column from the DataFrame.
# The 'axis=1' parameter indicates that a column is being dropped (use'axis=0' for rows).
# The 'inplace=True' parameter modifies the DataFrame directly without returning a new object.
if 'ColumnName' in df.columns:
    df.drop('ColumnName', axis=1, inplace=True)
else:
    print("'ColumnName' does not exist in the DataFrame.") # Removes the column named 'ColumnName'
# Rename columns
df.rename(columns={'OldName1': 'Past_Exam_Scores', 'OldName2': 'Final_Exam_Score'}, inplace=True)
# Assigning custom index (row) names
df.index = [f'Row{i+1}' for i in range(len(df))]

print(df)
```

6. Grouping Data

```python
# Assuming 'df' is your DataFrame
# Group by 'Gender' and calculate the mean for numeric columns
grouped_data = df.groupby('Gender').mean(numeric_only=True)

print(grouped_data)
```

7. Sorting Data

```python
import pandas as pd


# Sort the DataFrame by 'Final_Exam_Score' in ascending order
sorted_df = df.sort_values(by='Final_Exam_Score', ascending=True)

# Display the sorted DataFrame
print(sorted_df)
```
